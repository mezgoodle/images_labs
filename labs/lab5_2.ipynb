# Second task

```python
from imageai.Detection import VideoObjectDetection
```

## Create model

```python
detector = VideoObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath(common_path + 'yolo.h5')
detector.loadModel()
```

## Start the analysis

```python
video_path = detector.detectObjectsFromVideo(input_file_path=common_path + 'video.mp4', 
                                             output_file_path='video.mp4',
                                             frames_per_second=20, 
                                             log_progress=True)
```
