# First task

```python
from imageai.Prediction import ImagePrediction
```

## Create model

```python
prediction = ImagePrediction()
prediction.setModelTypeAsResNet()
prediction.setModelPath(common_path + "resnet50_weights_tf_dim_ordering_tf_kernels.h5")
prediction.loadModel()
```

```python
files = ['apple.jpg', 'banana.jpg', 'cherries.jpg', 'huckleberry.jpg', 'kiwi.jpg']
```

## Start the analysis

```python
for file in files:
  predicitions, percentage_probabilities = prediction.predictImage(common_path + 'Fruits/' + file, result_count=5)
  print(f'Назву файлу {file}')
  result = predicitions[0] == file.split('.')[0]
  print('Чи правильно виявлено фрукт?')
  print('Так' if result else 'Ні')
  print(f'Найперший варіант - {predicitions[0]} з відсотком {round(percentage_probabilities[0], 2)}')
```
